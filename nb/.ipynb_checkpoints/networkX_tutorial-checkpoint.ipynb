```python
import numpy as np
import networkx as nx
import matplotlib.pyplot as plt
```

# Undirected Graph

## Build Graph

```python
graph = nx.Graph()
```

```python
graph.clear()

graph.add_node(1, name=1)
graph.add_node(2, name=2)
graph.add_node('asdf', name='asdf')
graph.add_nodes_from([3, 4], name='jkl;')

graph.add_edge(1,2, grg='gram1')
graph.add_edge(1,3)
graph.add_edges_from([('asdf', 2), (1,4), ('asdf', 4)], grg='gram2')
```

## Plot Graph

```python
nx.draw(graph, with_labels=True)
plt.show()
```

```python
pos = nx.spring_layout(graph)
nx.draw(graph, pos)
node_labels = nx.get_node_attributes(graph, 'name')
nx.draw_networkx_labels(graph, pos, labels = node_labels)
edge_labels = nx.get_edge_attributes(graph,'grg')
nx.draw_networkx_edge_labels(graph, pos, labels = edge_labels)
plt.show()
```

## View Graph Properties

#### Nodes + Edges

```python
print(graph.nodes)
print(graph.edges)
```

```python
list(graph.adj)
```

# Directed Graph

```python
directed = nx.DiGraph()
```

```python
directed.clear()

graph.add_node(1, name=1)
graph.add_node(2, name=2)
graph.add_node('asdf', name='asdf')
graph.add_nodes_from([3, 4], name='jkl;')

directed.add_edge(1,2, grg='gram1')
directed.add_edge(1,3)
directed.add_edges_from([('asdf', 2), (1,4), ('asdf', 4)], grg='gram2')
```

```python
nx.draw(graph, with_labels=True)
plt.show()
```

```python
pos = nx.spring_layout(graph)
nx.draw(graph, pos)
node_labels = nx.get_node_attributes(graph, 'name')
nx.draw_networkx_labels(graph, pos, labels = node_labels)
edge_labels = nx.get_edge_attributes(graph,'grg')
nx.draw_networkx_edge_labels(graph, pos, labels = edge_labels)
plt.show()
```
